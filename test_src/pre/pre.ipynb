```python
import pandas as pd
from matplotlib import pyplot as plt

def process_data():
    # 加载数据
    df = pd.read_csv('/lustre/home/hangqian/projects/data/zircons/Roberts_Spencer_2015.csv')
    years = df["U-Pb (Ma)"]

    # 绘制直方图
    data = plt.hist(years, bins=200)
    zircons_data = data[0]
    ages = data[1][1:]

    return ages, zircons_data


if __name__ == "__main__":
    process_data()
```
